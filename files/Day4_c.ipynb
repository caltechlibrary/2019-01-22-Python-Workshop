```python
import plotnine as p9
import pandas as pd
```

```python
surveys_complete = pd.read_csv('data/surveys.csv')
surveys_complete = surveys_complete.dropna()
```

```python
plot = p9.ggplot(data=surveys_complete, 
                 mapping=p9.aes(x='weight',y='hindfoot_length'))
```

```python
plot = plot + p9.geom_point()
plot
```

```python
plot2 = p9.ggplot(data=surveys_complete, 
                 mapping=p9.aes(x='plot_id'))
plot2 = plot2 + p9.geom_bar()
plot2
```

```python
plot = p9.ggplot(data=surveys_complete, 
                 mapping=p9.aes(x='weight',
                                y='hindfoot_length',
                                fill='species_id',
                               color='species_id'))
plot = plot + p9.geom_point(alpha=0.1)
plot
```

```python
plot = plot + p9.xlab("Weight (g)")
plot
```

```python
plot = plot + p9.scale_x_log10()
plot
```

```python
plot = plot + p9.theme_bw()
plot = plot + p9.theme(text=p9.element_text(size=16))
plot
```

```python
plot2 = p9.ggplot(data=surveys_complete, 
                 mapping=p9.aes(x='plot_id',fill='sex'))
plot2 = plot2 + p9.geom_bar() +p9.scale_fill_manual(["blue","orange"])
plot2
```

```python
survey_subset = surveys_complete[surveys_complete["year"].isin([2000,2001])]
plot = p9.ggplot(data=survey_subset,mapping=p9.aes(x='weight',
                                                   y='hindfoot_length',
                                                   color='species_id'))
plot = plot + p9.geom_point(alpha=0.1)
plot = plot + p9.facet_grid("year ~ sex")
plot
```
