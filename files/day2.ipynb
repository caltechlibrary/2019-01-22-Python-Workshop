```python
def fahr_to_celsius(temp):
    return ((temp - 32)*(5/9))
```

```python
print('freezing point of water:',fahr_to_celsius(32), ' C')
```

```python
print('boiling point of water:',fahr_to_celsius(212), ' C')
```

```python
def celsius_to_kelvin(temp_c):
    return temp_c + 273.15
```

```python
def fahr_to_kelvin(temp_f):
    temp_c = fahr_to_celsius(temp_f)
    temp_k = celsius_to_kelvin(temp_c)
    return temp_k

print('boiling point of water in Kelvin:', fahr_to_kelvin(212.0))
```

```python
celsius_to_kelvin(fahr_to_celsius(212.0))
```

```python
import matplotlib.pyplot

def analyze(filename):
    
    data = numpy.loadtxt(fname=filename, delimiter=',')

    fig = matplotlib.pyplot.figure(figsize=(10.0,3.0))

    axes1 = fig.add_subplot(1,3,1)
    axes2 = fig.add_subplot(1,3,2)
    axes3 = fig.add_subplot(1,3,3)

    axes1.set_ylabel("average")
    axes1.plot(numpy.mean(data,axis=0))

    axes2.set_ylabel("min")
    axes2.plot(numpy.min(data,axis=0))

    axes3.set_ylabel("max")
    axes3.plot(numpy.max(data,axis=0))

    fig.tight_layout()

    matplotlib.pyplot.show()
```

```python
def detect_problems(f_name):
    data = numpy.loadtxt(fname=f_name, delimiter=',')
    if numpy.max(data, axis=0)[0] == 0 and numpy.max(data, axis=0)[20] == 20:
        print("Suspicous looking maxima")
    elif numpy.sum(numpy.min(data,axis=0)) == 0:
        print("Suspicous looking minima")
    else:
        print("OK")
```

```python
filenames = sorted(glob.glob('data/inflammation*.csv'))

for f in filenames:
    print(f)
    analyze(f)
    detect_problems(f)
```

```python
def offset_mean(data,target_mean_value=0):
    '''Return a new array containing the origional data
    with its mean offset to match the desired value.'''
    return(data-numpy.mean(data))+target_mean_value
```

```python
data = numpy.loadtxt(fname='data/inflammation-01.csv',delimiter=',')
print(data)
print(offset_mean(data))
```

```python
def normalize_rectangle(rect):
    '''Normalize a rectangle so that its 
    at the origin and 1.0 units long on its longest axis.
    Input should be of the format (x0,y0,x1,y1)
    With x0,y0 the lower left and y1,y1 the upper right corners'''
    assert len(rect) == 4, 'Rectangle must contain 4 coordinates'
    x0,y0,x1,y1 = rect
    assert x0 < x1, 'Invalid X coordinates'
    assert y0 < y1, 'Invalid Y coordinates'
    
    dx = x1 - x0
    dy = y1 - y0
    if dx > dy:
        scaled = float(dx) / dy
        upper_x, upper_y = 1.0, scaled
    else:
        scaled = float(dx) / dy
        upper_x, upper_y = scaled, 1.0
        
    assert 0 < upper_x <= 1.0, 'Calculated upper X coordinate invalid'
    assert 0 < upper_y <= 1.0, ' Calculated upper Y coordinate invalid'
    
    return(0.0,0.0, upper_x,upper_y)
```

```python
print(normalize_rectangle((0.0,0.0,1.0,5.0)))
```

```python
print(normalize_rectangle((0.0,0.0,5.0,1.0)))
```
